Remove outliers, null values, and observations with a value of -1

```python
import pandas as pd
import numpy as np
import warnings
import csv
import os
from os import path
```

```python
filename = os.listdir(r'.\Pre')
```

```python
#数据清洗
#创建循环，遍历每一个csv文件
for i in range(len(filename)):
    #读取文件
    df = pd.read_csv(r'./Pre/'+ filename[i])
    
    #删除有空值的行
    df = df[df['eyeOpenLeft'].notna()]
    df = df[df['eyeOpenRight'].notna()]
    df = df[df['pupilDiameterLeft'].notna()]
    df = df[df['pupilDiameterRight'].notna()]
    df = df[df['time'].notna()]
    
        
    #删除“-1”
    df = df[(df["eyeOpenLeft"] != -1) & (df["eyeOpenRight"] != -1)]
    df = df[(df["pupilDiameterLeft"] > 0) & (df["pupilDiameterRight"]>0)]
    
    #删除离群点
    q_low_1 = df["eyeOpenLeft"].quantile(0.01)
    q_hi_1  = df["eyeOpenLeft"].quantile(0.99)

    q_low_2 = df["eyeOpenRight"].quantile(0.01)
    q_hi_2  = df["eyeOpenRight"].quantile(0.99)

    q_low_3 = df["pupilDiameterLeft"].quantile(0.01)
    q_hi_3  = df["pupilDiameterLeft"].quantile(0.99)

    q_low_4 = df["pupilDiameterRight"].quantile(0.01)
    q_hi_4  = df["pupilDiameterRight"].quantile(0.99)

    cf = df[((df["eyeOpenLeft"] <= q_hi_1) & (df["eyeOpenLeft"] >= q_low_1)) & 
    ((df["eyeOpenRight"] <= q_hi_2) & (df["eyeOpenRight"] >= q_low_2)) &
    ((df["pupilDiameterLeft"] <= q_hi_3) & (df["pupilDiameterLeft"] >= q_low_3)) & 
    ((df["pupilDiameterRight"] <= q_hi_4) & (df["pupilDiameterRight"] >= q_low_4))]
    
    #加列名
    cf.columns = ['eyeOpenLeft', 'eyeOpenRight', 'pupilDiameterLeft',"pupilDiameterRight","time"]

    #将处理好的数据导出
    cf.to_csv(r'./newPre/' + filename[i],mode='a')
    #print(filename[i]+'********文件处理成功')
```

```python
#ROCindex(post - pre/ pre)
for item in urllist:
    rocurl = path.join(item, 'after\mydata.txt')
    with open(rocurl) as f:
        allline = []
        for line in f:
            if line == 'miss\n' or 'miss':
                allline.append(0.8)
            else:
                allline.append(float(line))
        postavg = sum(allline) / len(allline)
    rocurl = path.join(item, 'pre\mydata.txt')
    with open(rocurl) as f:
        allline = []
        for line in f:
            if line == 'miss\n':
                allline.append(0.8)
            else:
                allline.append(float(line))
        preavg = sum(allline) / len(allline)
    roclist.append((postavg - preavg) / preavg)
    roclist.append((postavg - preavg) / preavg)
    roclist.append((postavg - preavg) / preavg)
```
